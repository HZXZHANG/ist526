```python
# python visualization libraries
import pandas as pd
import numpy as np

import math
import json


import plotly.express as px
import plotly.graph_objects as go

# for hierarchical data
import networkx as nx
```

```python
line = {'color': "black", 'width': 0.5}
print(line)

node = {'pad': 15, 
        'thickness': 20, 
        'line': line,
        'label': ["Automotive Sales", "Automotive Leasing","Automotive Regulatory Credits", "Energy Generation and Storage", 
                  "Services and Other", "Total Revenue","Gross Profit","Interest Income",
                  "Net Income","Interest Expense","Operating Income","Tax & Other Expense", 
                  "Operating Expenses","SG&A", "R&D","Cost of Revenue"],
        'color': "blue"
      }
print(node)

link = {
      'source': [ 0, 1, 2, 3, 4, 5, 5, 6, 6,10,10,10, 7,12,12], # indices correspond to labels, (e.g., A1=0, A2=1) and (souce_i, target_i) are tuple  
      'target': [ 5, 5, 5, 5, 5, 6,15,10,12, 8, 9,11, 8,13,14],
      'value' : [15, 1, 1, 2, 3, 4, 18, 4, 2, 4, 1, 1, 1, 1, 1]
    }
print(link)
```

```python
fig = go.Figure(
  data = [go.Sankey(node = node, link = link)]
)

fig.update_layout(
  title_text="Basic Sankey Diagram", 
  font_size=10
)
fig.show()
```
