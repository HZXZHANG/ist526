```python
# ref: https://stackoverflow.com/questions/32400867/pandas-read-csv-from-url

url = '/content/drive/MyDrive/Colab Notebooks/data/hierarchical data.csv'

# pandas call to read csv file 
df = pd.read_csv(url)

# quickly show the dataframe
df.head()
```

```python
# remove NaN with blank, otherwise plotly will be upset
df.fillna('', inplace = True)
# df.dropna(axis=0, inplace = True)
display(df.head())
```

```python
fig = px.treemap(
  df,   
  parents = 'Parent',
  names = 'Item and Group',
  values='Weight'
  #color='Indent Level'
)

fig.update_layout(
  title_text="Treemap Diagram", 
  font_size=12
)

fig.show()
```
